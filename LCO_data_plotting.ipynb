```
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
```

```
# Reading in the B filter and V filter catalogues from the open clusters

# COIN Gaia 15
B_mag, B_mag_err, V_mag, V_mag_err, BminusV = np.loadtxt('BV_cat.txt', unpack = True, skiprows = 1, usecols = [7,8,25,26,37], delimiter = ',', dtype = float)

# Converting Apparent magnitudes to Absolute magnitudes
dist_pc = 1174
Vmag_ab = V_mag - 5 * np.log10(dist_pc) + 5
Bmag_ab = B_mag - 5 * np.log10(dist_pc) + 5

print(Vmag_ab - Bmag_ab)
print(BminusV)

# COIN Gaia 27
B1_mag, B1_mag_err, V1_mag, V1_mag_err, BminusV1 = np.loadtxt('BV1_cat.txt', unpack = True, skiprows = 1, usecols = [7,8,25,26,37], delimiter = ',', dtype = float)

# Converting Apparent magnitudes to Absolute magnitudes
dist_pc = 1173
V1mag_ab = V1_mag - 5 * np.log10(dist_pc) + 5
B1mag_ab = B1_mag - 5 * np.log10(dist_pc) + 5
```

```
# COIN Gaia 15

# Error for the BV scale
BV_Errors = np.sqrt((B_mag_err)**2 + (V_mag_err)**2)

# Plotting data and error bars
figure(figsize= (10,10))
plot(BminusV, Vmag_ab, "k.", markersize=1)
plt.errorbar(BminusV, Vmag_ab, xerr=BV_Errors, yerr=V_mag_err, alpha=0.1, fmt="none", ecolor="red")
plt.gca().invert_yaxis()

ylabel("Absolute Magnetude of V band")
xlabel("B-V colour scale")
title('H-R diagram for COIN Gaia 15')

grid();
```

```
# COIN Gaia 27

# Error for the BV scale
BV1_Errors = np.sqrt((B1_mag_err)**2 + (V1_mag_err)**2)

# Plotting data and error bars
figure(figsize= (10,10))
plot(BminusV1, V1mag_ab, "k.", markersize=1)
plt.errorbar(BminusV1, V1mag_ab, xerr=BV1_Errors, yerr=V1_mag_err, alpha=0.1, fmt="none", ecolor="red")
plt.gca().invert_yaxis()

ylabel("Absolute Magnetude of V band")
xlabel("B-V colour scale")
title('H-R diagram for COIN Gaia 27')

grid();
```

```
# Using the Ancillary data

# COIN Gaia 15
# Reading in the text file
Gmag, GbpGrp_mag = np.loadtxt('CG15.txt', unpack = True, usecols = [23,24], dtype = float)
dist_pc = 1174
Gmag_ab = Gmag - 5 * np.log10(dist_pc) + 5

# COIN Gaia 27
# Reading in the text file
G1mag, G1bpG1rp_mag = np.loadtxt('CG27.txt', unpack = True, usecols = [23,24], dtype = float)
dist1_pc = 1173
G1mag_ab = G1mag - 5 * np.log10(dist1_pc) + 5
```

```
# Gathering and sorting Isochrone data

# Age varying 1e6 to 1e7 in steps of 1e6 with metallicity of Z = 0.003
iso_logAgeAD, iso_Gmag, iso_GBpmagAD, iso_GRpmagAD = np.loadtxt('AncillaryIso.txt', unpack = True, usecols = [2, 28, 29, 30], dtype = float)
print(np.unique(iso_logAgeAD))

# Age varying 1e7 to 1e8 in steps of 1e7 with metallicity of Z = 0.003
iso_logAgeAD1, iso_Gmag1, iso_GBpmagAD1, iso_GRpmagAD1 = np.loadtxt('AncillaryIso1.txt', unpack = True, usecols = [2, 28, 29, 30], dtype = float)
print(np.unique(iso_logAgeAD1))

# Age varying 1e8 to 1e9 in steps of 1e8 with metallicity of Z = 0.003
iso_logAgeAD2, iso_Gmag2, iso_GBpmagAD2, iso_GRpmagAD2 = np.loadtxt('AncillaryIso2.txt', unpack = True, usecols = [2, 28, 29, 30], dtype = float)
print(np.unique(iso_logAgeAD2))
```

```
# Sorting into arrays which match the desired data
age1 = np.where(iso_logAgeAD == 6.47712)
age2 = np.where(iso_logAgeAD == 6.90309)
age3 = np.where(iso_logAgeAD1 == 7.47712)
age4 = np.where(iso_logAgeAD1 == 7.90309)
age5 = np.where(iso_logAgeAD2 == 8.47712)
age6 = np.where(iso_logAgeAD2 == 8.90309)

# Isochrone magnitudes for age 1
Gmag_age1 = iso_Gmag[age1]
GBpmag_age1 = iso_GBpmagAD[age1]
GRpmag_age1 = iso_GRpmagAD[age1]
GBpminGRp_age1 = GBpmag_age1 - GRpmag_age1

# Isochrone magnitudes for age 2
Gmag_age2 = iso_Gmag[age2]
GBpmag_age2 = iso_GBpmagAD[age2]
GRpmag_age2 = iso_GRpmagAD[age2]
GBpminGRp_age2 = GBpmag_age2 - GRpmag_age2

# Isochrone magnitudes for age 3
Gmag_age3 = iso_Gmag1[age3]
GBpmag_age3 = iso_GBpmagAD1[age3]
GRpmag_age3 = iso_GRpmagAD1[age3]
GBpminGRp_age3 = GBpmag_age3 - GRpmag_age3

# Isochrone magnitudes for age 4
Gmag_age4 = iso_Gmag1[age4]
GBpmag_age4 = iso_GBpmagAD1[age4]
GRpmag_age4 = iso_GRpmagAD1[age4]
GBpminGRp_age4 = GBpmag_age4 - GRpmag_age4

# Isochrone magnitudes for age 5
Gmag_age5 = iso_Gmag2[age5]
GBpmag_age5 = iso_GBpmagAD2[age5]
GRpmag_age5 = iso_GRpmagAD2[age5]
GBpminGRp_age5 = GBpmag_age5 - GRpmag_age5

# Isochrone magnitudes for age 6
Gmag_age6 = iso_Gmag2[age6]
GBpmag_age6 = iso_GBpmagAD2[age6]
GRpmag_age6 = iso_GRpmagAD2[age6]
GBpminGRp_age6 = GBpmag_age6 - GRpmag_age6
```

```
# Creating colour-colour magnitude diagrams using the ancillary data in the G band filter

# COIN Gaia 15

# Plotting data and error bars
figure(figsize= (10,10))
plot(GbpGrp_mag, Gmag_ab, "k.", markersize=1)
plt.gca().invert_yaxis()

ylabel("Absolute Magnetude of G band")
xlabel("G(Bp) - G(Rp) colour scale")
#title('COIN Gaia 15, isochone model using Gaia DR2 data')

ylim(7.5,0)
xlim(-0.5,2)

plot(GBpminGRp_age1, Gmag_age1, 'r.', label = '10$^{6.47712}$ years')
plot(GBpminGRp_age2, Gmag_age2, 'y.', label = '10$^{6.90309}$ years')
plot(GBpminGRp_age3, Gmag_age3, 'g.', label = '10$^{7.47712}$ years')
plot(GBpminGRp_age4, Gmag_age4, 'b.', label = '10$^{7.90309}$ years')
plot(GBpminGRp_age5, Gmag_age5, 'm.', label = '10$^{8.47712}$ years')
plot(GBpminGRp_age6, Gmag_age6, 'c.', label = '10$^{8.90309}$ years')

legend(fontsize = 'large')

grid();
```
